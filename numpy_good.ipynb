```python
import numpy as np
```

```python
#input by user


persons = int(input("Enter number of persons: "))
days = int(input("Enter number of days: "))

attendance = []

print("\nEnter attendance (1 = Present, 0 = Absent):")
for i in range(persons):
    data = input(f"Person {i+1}: ")
    row = [int(x) for x in data]   # convert each digit
    attendance.append(row)

attendance = np.array(attendance)
```

```python
#formula
percentages = attendance.sum(axis=1) / days * 100
```

```python

print("\nAttendance Percentage:")
for i, p in enumerate(percentages, start=1):
    print(f"Person {i}: {p:.2f}%")

print("\nLow Attendance (<75%):")
for i, p in enumerate(percentages, start=1):
    status = "YES" if p < 75 else "NO"
    print(f"Person {i}: {status}")
```
